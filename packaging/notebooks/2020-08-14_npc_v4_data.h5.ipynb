```python
# I expect the number of NaNs in concatted synth to be 
# size of concatted - sum(sizes of all the sessions)
print(np.count_nonzero(np.isnan(concatted)))
print(concatted.size)
print(sum([bk19_assys[k].size for k in synth_names]))
print(concatted.size - sum([bk19_assys[k].size for k in synth_names]))
assert np.count_nonzero(np.isnan(concatted)) == concatted.size - sum([bk19_assys[k].size for k in synth_names])
# or
# sum(shape[0] of all synth assys) * sum(shape[1] of all synth assys) - sum(sizes of all synth assys)
assert np.count_nonzero(np.isnan(concatted)) == (lambda x: sum(x[0]) * sum(x[1]))(list(zip(*[bk19_assys[k].shape for k in synth_names]))) - sum([bk19_assys[k].size for k in synth_names])
```

```python
# I expect the number of NaNs in concatted naturalistic to be 
# max(shape[0] of all nat assys) * sum(shape[1] of all nat assys) - sum(size of all nat assys)
print(max([bk19_assys[k].shape[0] for k in nat_names]) * sum(bk19_assys[k].shape[1] for k in nat_names))
print(sum(bk19_assys[k].size for k in nat_names))
print(max([bk19_assys[k].shape[0] for k in nat_names]) * sum(bk19_assys[k].shape[1] for k in nat_names) - sum(bk19_assys[k].size for k in nat_names))
print(np.count_nonzero(np.isnan(concatted_nat)))
assert np.count_nonzero(np.isnan(concatted_nat)) == max([bk19_assys[k].shape[0] for k in nat_names]) * sum(bk19_assys[k].shape[1] for k in nat_names) - sum(bk19_assys[k].size for k in nat_names)
```
