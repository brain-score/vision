```python
def showme(thingy, depth=0, name=None):
    indent = f"{'    '*depth}"
    result = indent
    if name:
        result = result + f"{name}:  "
    result = result + f"{type(thingy).__name__} "
    try:
        dtype = thingy.dtype
        shape = thingy.shape
        result = result + f"{str(shape)} {str(dtype)} "
        if dtype == "O":
            if shape == ():
                result = result + ":  \n" + showme(thingy[()], depth=depth+1)
            else:
                raise TypeError
        return result + "\n"
    except TypeError:
        pass
    except AttributeError:
        pass
    try:
        keys = thingy.keys()
        result = result + f"{str(len(thingy))} "
        result = result + "{\n"
        for key in keys:
            result = result + showme(thingy[key], depth=depth+1, name=key)
        result = result + indent + "}\n"
        return result
    except AttributeError:
        pass
    try:
        iterator = iter(thingy)
        result = result + f"{str(len(thingy))} "
        result = result + "[\n"
        for thing in iterator:
            result = result + showme(thing, depth=depth+1)
        result = result + indent + "]\n"
        return result
    except TypeError:
        pass
    return result + "\n"
```

```python
def get_image_ids(variation, data):
    ids = df_hvm[df_hvm["var"]==f"V{variation}"]["id"]
    assert data.shape[0] == len(ids)
    return ids
```

```python
def coords_from_darren(animals_index, animal_region, variation, data, xr_data=None):
    coords = {}
    coords["image_id"] = ("stimulus", get_image_ids(variation, data))
    coords["repetition"] = ("repetition", range(data.shape[1]))
    coords["animal"] = ("neuroid", [animal_region.split("_")[0]]*data.shape[2])
    coords["region"] = ("neuroid", [animal_region.split("_")[1]]*data.shape[2])
    coords["electrode"] = ("neuroid", range(data.shape[2]))
    coords["neuroid_id"] = ("neuroid", [f"{animal_region}_{i:02}" for i in range(data.shape[2])])
    return coords
```

```python
coords_from_darren(**relevant_arrays[0])
```

```python
def xr_from_darren(animals_index, animal_region, variation, data, xr_data=None):
    xr_data = xr.DataArray(
        data=data,
        coords=coords_from_darren(animals_index, animal_region, variation, data),
        dims=["stimulus", "repetition", "neuroid"]
    )
    return xr_data
```

```python
for ra in relevant_arrays:
    ra["xr_data"] = xr_from_darren(**ra)
relevant_arrays
```

```python
concatted_per_var = {}
for v in var:
    aligned = xr.align(*[x["xr_data"] for x in relevant_arrays if x["variation"]==v], 
         join="outer", 
         exclude=["neuroid"], 
         fill_value=np.nan
        )
    concatted_per_var[v] = xr.concat(aligned, dim="neuroid")
concatted_per_var
```

```python
stacked_per_var = {}
for v in concatted_per_var:
    stacked_per_var[v] = concatted_per_var[v].stack(presentation=["repetition", "stimulus"])
stacked_per_var
```
